```python
from random import * # 1. 1~50번째 손님을 반복
count = 0
for customer in range(1, 51):
    # 2. 5분 ~ 50분 사이의 소요 시간을 무작위로 생성
    time = randint(5, 50)
    
    # 3. 소요 시간이 5분 ~ 15분 사이인 손님만 [O] 표시
    # (기존 코드의 조건 16 < time은 17분부터 해당하므로 조건에 맞춰 수정)
    if 4 < time < 16:
        print("[O] {0}번째 손님 (소요시간 : {1}분)".format(customer, time))
        count += 1 #총 탑승 승객 카운트
    else:
        print("[ ] {0}번째 손님 (소요시간 : {1}분)".format(customer, time))
print("총 탑승 승객 : {0} 분".format(count))
    
```

```python
#지역 변수, 전역 변수
gun = 10

def checkpoint(soldiers):
    global gun #전역 공간에 있는 gun사용
    gun = gun - soldiers
    print("[] last gun: {0}".format(gun))

def checkpoint_ret(gun, soldiers):
    gun = gun - soldiers
    print("[] last gun: {0}".format(gun))
    return gun
    
print("All guns : {0}".format(gun))
checkpoint(2)
print("All guns : {0}".format(gun))
gun = checkpoint_ret(gun, 2)
print("All guns : {0}".format(gun))
```

```python
#문제
def std_weight(height, gender):
    if gender == "male":
        print("{0}cm 남자의 표준 체중은 {1:.2f}kg 입니다.".format(height, height*height*22*0.0001))
    elif gender == "female":
        print("{0}cm 여자의 표준 체중은 {1:.2f}kg 입니다.".format(height, height*height*210*0.0001))
    else:
        print("Wrong answer.")
        
kilogram = std_weight(175, "male")
print(kilogram)
```
