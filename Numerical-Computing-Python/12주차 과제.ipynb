```python
st_sc = {
    "홍": {"국어": 90, "영어": 85, "수학": 92},
    "김": {"국어": 75, "영어": 80, "수학": 78},
    "이": {"국어": 88, "영어": 92, "수학": 95}
}

filename = "st_sc.csv"

h= "이름.국어.영어.수학\n"
rows = ""

for name, scores in st_sc.items():
    row = f"{name},{scores['국어']},{scores['영어']},{scores['수학']}\n"
    rows += row

with open(filename, "w", encoding="utf8") as f:
    f.write(h)
    f.write(rows)

print(f"{filename} 파일로 저장 완료")
```

```python
#17 중 (2점)

def star_print_file(file, num):
    if num == 1:
        file.write("*\n")
        return
    else:
        star_print_file(file, num-1)
        file.write("*" * num + "\n")

with open("별찍기.txt", "wt") as f:
    # 파라미터 전달
    star_print_file(f, 10)
    print("별찍기 결과 저장 완료")
    f.close()

with open("별찍기.txt", "rt") as f:
    for line in f:
        print(line, end="")
    f.close()
```

```python
#20 중 (2점)

import time

start_time = time.time()
elapsed = 0

while True:
    user_input = input(f"[{elapsed:.2f}초] s = 기록, q = 종료 > ")

    if user_input.lower() == "q":
        break
    elif user_input.lower() == "s":
        try:
            now = time.time()
            elapsed = now - start_time
            with open("laps.txt", "a", encoding="utf-8") as file:
                file.write(f"{elapsed:.2f}초\n")
        except IOError:
            print("오류: 파일에 기록할 수 없습니다.")
        else:
            print("기록되었습니다.")

print("스톱워치가 종료되었습니다.")
```

```python
#21 중 (2점)

grade_map = [(90, "A"), (80, "B"), (70, "C"), (60, "D")]

def get_grade(score):
    for threshold, grade in grade_map:
        if score >= threshold:
            return grade
    return "F"

results = []
try:
    with open("scores.txt", "r", encoding="utf-8") as f:
        for line in f:
            name, score = line.strip().split(",")
            results.append((name, int(score), get_grade(int(score))))
except FileNotFoundError:
    print("오류: scores.txt 파일을 찾을 수 없습니다.")
else:
    # 쓰기 부분에 encoding="utf-8" 추가
    with open("results.txt", "w", encoding="utf-8") as f:
        f.write("이름,점수,등급\n")
        for name, score, grade in results:
            f.write(f"{name},{score},{grade}\n")
    print("결과가 results.txt에 저장되었습니다.")
    for name, score, grade in results:
        print(f"{name}: {score}점 → {grade}등급")
```

```python
#22 중 (2점)

frequency = {}

with open("words.txt", "r", encoding="utf-8") as f:
    words = f.read().lower().split()
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

sorted_words = sorted(frequency.items(), key=lambda x: x[1], reverse=True)

with open("top_words.txt", "w", encoding="utf-8") as f:
    f.write("순위,단어,빈도수\n")
    for rank, (word, count) in enumerate(sorted_words[:3], start=1):
        line = f"{rank},{word},{count}\n"
        f.write(line)
        print(line.strip())
```
