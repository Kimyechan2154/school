```python
import time
print("# time.time()")
t = time.time()
print("타임스탬프:", t)

print("# time.localtime()")
lt = time.localtime()
print("현지:", lt)

print("# time.localtime()")
s = time.strftime("%Y-%m-%d %H:%M:%S", lt)
print("포멧:", s)

print("# time.strftime()")
s = time.strftime("%Y-%m-%d %H:%M:%S", lt)
print("포멧:", s)

print("# time.strptime()")
ts = time.strptime("2025-05-11 15:30:00","%Y-%m-%d %H:%M:%S")
print("파싱:", ts)
print(f"{ts.tm_year}년 {ts.tm_mon}년 {ts.tm_mday}년")

print("# time.sleep(5)")
ts = time.sleep(5)
print("파싱:", ts)
```

```python
def fac(n):
    if n == 0:
        return 1
    else:
        return n*fac(n-1)

print(fac(5))
```

```python
def fibo(n):
    if n < 2:
        return n
    else:
        return fibo(n-2) + fibo(n-1)

print(fibo(5))
```

```python
class Person:                          # ① 클래스 선언 (설계도)
    def __init__(self, name, age):     # ② 생성자: 객체 생성 시 초기값 설정
        self.name = name               #   멤버변수: 이름
        self.age = age                 #   멤버변수: 나이
        
    def introduce(self):               # ③ 메소드: 자기소개 출력
        print(f"이름은 {self.name}, 나이는 {self.age}살")

    def birthday(self):                # ④ 메소드: 나이 1 증가 후 출력
        self.age += 1
        print(f"{self.name}의 생일! 이제 {self.age}살.")

p1 = Person("hkkim", 20)   # ← 여기서 객체(인스턴스)가 만들어짐
p1.introduce()             # 이름은 hkkim, 나이는 20살
del
```

```python
string_list = ["1", "2", "3", "4", "a", "b", "c"]
# 문자열 리스트에서 숫자형인 데이터만 필터링
num_string_filter = filter(str.isdecimal, string_list)
print(num_string_filter, type(num_string_filter))
num_string_list = list(num_string_filter)

print('')
print(num_string_list, type(num_string_list))
# String형으로 저장되어 있는 숫자들을 int형으로 일괄 변경

print('')
int_map = map(int, num_string_list)
print(int_map, type(int_map))
int_list = list(int_map)

print('')
print("{} -> {} -> {}" .format(string_list ,list(num_string_list), list(int_list)))
```

```python
string_converter = lambda parameter: str(parameter)
num_list = [1, 2, 3, 4, 5]
convert_result = string_converter(num_list)
print("{} {} -> {} {}" .format(num_list, type(num_list), convert_result, type(convert_result)))
for x in num_list:
 convert_result = string_converter(x)
 print("{} {} -> {} {}" .format(x, type(x), convert_result, type(convert_result)))
```

```python
memo = {
 1: 1,
 2: 2
}
call_counter = 0
def fibonacci_memo(n):
 global call_counter
 call_counter += 1
 if n in memo:
     return memo[n]
 else:
     output = fibonacci_memo(n - 1) + fibonacci_memo(n - 2)
     memo[n] = output
     return output
import time
# 함수 사용 전, 시간 체크
start = time.time()
call_counter = 0; print("fibonacci_memo(14): {}, 함수 깊이: {}" .format(fibonacci_memo(14), call_counter))
call_counter = 0; print("fibonacci_memo(30): {}, 함수 깊이: {}" .format(fibonacci_memo(30), call_counter))
call_counter = 0; print("fibonacci_memo(35): {}, 함수 깊이: {}" .format(fibonacci_memo(35), call_counter))
# 함수 사용 후, 시간 체크
end = time.time()
print("전체 실행시간", end-start)
```
